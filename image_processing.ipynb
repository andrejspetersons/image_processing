```python
#090 aptumšošana,reizināšana
import numpy as np
from numpy import asarray# lai apstrādātu un piekļūtu pie attēla pikseļiem
import cv2#OpenCv
from PIL import Image#attēlu apstrādei,glabāšanai
import matplotlib.pyplot as plt
import matplotlib.image as mpimg#attēlu apstrādei,glabāšanai

imageA=cv2.imread('2.jpg')
imageB=cv2.imread('mustang.jpg')

im_rgbA=cv2.cvtColor(imageA,cv2.COLOR_BGR2RGB)
im_rgbB=cv2.cvtColor(imageB,cv2.COLOR_BGR2RGB)

resizedA=cv2.resize(imageA,(500,500))
resizedB=cv2.resize(imageB,(500,500))

cv2.imwrite('mustang.jpg',resizedA)
cv2.imwrite('2.jpg',resizedB)

image_A=Image.open('mustang.jpg')
image_B=Image.open('2.jpg')

data_A=asarray(image_A)
data_B=asarray(image_B)

image_array_A=Image.fromarray(data_A)
image_array_B=Image.fromarray(data_B)

image_normalize_A=data_A/255.0
image_normalize_B=data_B/255.0




#darken numpy.where(conditions,True,False)
darken=np.where(image_normalize_A<=image_normalize_B,image_normalize_A,image_normalize_B)

fig=plt.figure(figsize=(10,6),dpi=100,facecolor='w',edgecolor='k')


ax1=fig.add_subplot(1,4,1)
ax1.title.set_text("picture1")
ax1.imshow(image_normalize_A)

ax2=fig.add_subplot(1,4,2)
ax2.title.set_text("picture2")
ax2.imshow(image_normalize_B)

ax3=fig.add_subplot(1,4,3)
ax3.title.set_text("darken")
ax3.imshow(darken)

#multiply
#C=A*B

multiply=image_normalize_A*image_normalize_B

fig=plt.figure(figsize=(10,6),dpi=100,facecolor='w',edgecolor='k')

ax1=fig.add_subplot(1,4,1)
ax1.title.set_text("picture1")
ax1.imshow(image_normalize_A)

ax2=fig.add_subplot(1,4,2)
ax2.title.set_text("picture2")
ax2.imshow(image_normalize_B)

ax3=fig.add_subplot(1,4,3)
ax3.title.set_text("multiply")
ax3.imshow(multiply)

#addition
#C=A+B



addition=image_normalize_A+image_normalize_B
addition=np.clip(addition,0,1)

fig=plt.figure(figsize=(10,6),dpi=100,facecolor='w',edgecolor='k')

ax1=fig.add_subplot(1,4,1)
ax1.title.set_text("picture1")
ax1.imshow(image_normalize_A)

ax2=fig.add_subplot(1,4,2)
ax2.title.set_text("picture2")
ax2.imshow(image_normalize_B)

ax3=fig.add_subplot(1,4,3)
ax3.title.set_text("addition")
ax3.imshow(addition)





```
